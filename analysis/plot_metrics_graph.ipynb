```python
from matplotlib.pylab import plt
import pandas as pd
from collections import defaultdict
%matplotlib inline 
```

```python
complete_datasets = ['sofmattress', 'powerplay11', 'curekart']
subset_datasets = ['sofmattress_subset', 'powerplay11_subset', 'curekart_subset']
models = ['dialogflow', 'rasa', 'luis', 'haptik', 'bert']
colors = ['orange', 'red', 'green', 'blue', 'pink']
```

```python
dataset = 'sofmattress'
for i, m in enumerate(models):
    df_res = pd.read_csv(f'../results/{m}_{dataset}.csv')
    df_res_subset = pd.read_csv(f'../results/{m}_{dataset}_subset.csv')
    mcc = list(df_res['MCC'])
    mcc_subset = list(df_res_subset['MCC'])
    threshold = list(df_res['Threshold'])
    plt.plot(threshold, mcc, '-', label=m, color=colors[i])
    plt.plot(threshold, mcc_subset, '--', color=colors[i])
    
legend1 = plt.legend(["Full", "Subset"], bbox_to_anchor=(1, 0.9))
legend1.legendHandles[0].set_color('black')
legend1.legendHandles[1].set_color('black')
plt.legend(loc="lower center")

plt.title("SOFMattress", fontsize=14)
plt.xlabel("Threshold")
plt.ylabel("Matthew's Correlation Coefficient")
plt.gca().add_artist(legend1)

plt.show()
```

```python
# In-scope accuracy
scores = []
for m in models:
    for d in complete_datasets+subset_datasets:
        df = pd.read_csv(f'../results/{m}_{d}.csv')
        scores.append((m,d,df[df['Threshold']==0.1]['Inscope Accuracy'].iloc[0]))
mydf = pd.DataFrame(scores, columns=['Platform', 'Dataset', 'Inscope Accuracy'])
mydf.pivot(index='Platform', columns='Dataset', values='Inscope Accuracy')
```
